## 선 색상과 스타일

```python
plt.plot(x, np.sin(x - 0), color='blue')        # 이름
plt.plot(x, np.sin(x - 1), color='g')           # 코드 (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # 0 and 1 사이 회색조
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # 16진수 코드 (RRGGBB, 00 ~ FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB 튜플, 0 ~ 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # 모든 HTML 색상 이름을 지원
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# 축약형 코드
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

## 제목, 라벨, 범례 표시

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

## 객체 지향 인터페이스: ax.set()메서드

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```

## 밀도 플롯과 등고선 플롯(Density, Contour)

```python
plt.rcParams.update(plt.rcParamsDefault)

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar()
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```
